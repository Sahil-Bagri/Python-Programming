# **`Seaborn`**

1. Provides a layer of abstraction hence simpler to use
2. More graphs included
3. Data visualization library based on matplotlib
4. Provides a high-level interface for drawing attractive
5. Informative statistical graphics.

#Types of Functions:
1. Figure Level
2. Axis Level

#Classification:
1.Relational Plot
2.Distribution Plot
3.Categorical Plot

**Relational Plot- Scatter Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#Scatter plot- axis level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#rel plot- figure level function-square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
#seaborn.lineplot
#The flights dataset has 10 years of monthly airline passenger data.
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'Jun'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
flights_wide = flights.pivot(index="month", columns="year", values="passengers")
flights_wide.head()
```

```python
sns.lineplot(data=flights_wide[1952])
```

```python
sns.lineplot(data=flights_wide)
```

```python
#seaborn.displot
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(data=penguins, x="flipper_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
```

```python
#seaborn.histplot
#A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm")
```

```python
sns.histplot(data=penguins, y="flipper_length_mm")
```

```python
sns.histplot(data=penguins, y="flipper_length_mm", bins=30)
```

```python
#seaborn.kdeplot
#Plot univariate or bivariate distributions using kernel density estimation.
#A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.
#KDE represents the data using a continuous probability density curve in one or more dimensions.
```

```python
#Plot a univariate distribution along the x axis:
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
```

```python
sns.kdeplot(data=tips, y="total_bill")
```

```python
iris = sns.load_dataset("iris")
sns.kdeplot(data=iris)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```python
#seaborn.boxplot
#A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.df = sns.load_dataset("titanic")
titanic=sns.load_dataset("titanic")
titanic

sns.boxplot(x=titanic["age"])
```

```python
sns.boxplot(data=titanic, x="class", y="age", hue="alive")
```

```python
#seaborn.violinplot
#It shows the distribution of data points after grouping by one (or more) variables.
sns.violinplot(x=titanic["age"])
```

```python
sns.violinplot(data=titanic, x="age", y="class")
```

```python
sns.violinplot(data=titanic, x="class", y="age", hue="alive")
```

```python
#seaborn.barplot
#A bar plot represents an aggregate or statistical estimate for a numeric variable with the height of each rectangle and indicates the uncertainty around that estimate using an error bar.
penguins = sns.load_dataset("penguins")
sns.barplot(penguins, x="island", y="body_mass_g")
```

```python
sns.barplot(penguins, x="body_mass_g", y="island", hue="island", legend=True)
```

```python
#seaborn.countplot
#A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.
sns.countplot(data=titanic, x="class")
```

```python
sns.countplot(data=titanic, x="class", hue="survived")
```

```python
sns.countplot(data=titanic, x="class", hue="survived", stat="percent")
```

```python
#seaborn.heatmap
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
sns.heatmap(glue)
```

```python
sns.heatmap(glue, annot=True)
```

```python
#seaborn.pairplot
#A univariate distribution plot is drawn to show the marginal distribution of the data in each column.
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins, hue="species")
```
