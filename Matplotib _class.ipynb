# Matplotlib:

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source.

```python
#Matplotlib Labels and Title
#Create Labels for a Plot
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Nitin")
plt.show()
```

## Matplotlib Scatter
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,color="red")
plt.show()

```python
#Matplotlib Bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x=np.array(["F","A","S","C"])
y=np.array([34,56,113,79])
color=["red","blue","pink","green"]
widht=[0.5,0.3,0.8,0.7]
plt.bar(x,y,width=widht,color=color,label="Arm")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([12,32,5,2])
width=[0.7,0.9,0.5,0.3]
plt.bar(x,y,color="green",width=width)
plt.grid()
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
color=["lightblue","red","blue","pink"]
plt.bar(x, y, color = color)
plt.show()
```

```python
x=np.array(["Sam","Vipin","Rahul","Bhavesh","Aman"])
y=np.array([23,34,54,8,14])
plt.pie(y,autopct="%1.f%%",startangle=180,labels=x)
plt.legend(loc=4)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```
